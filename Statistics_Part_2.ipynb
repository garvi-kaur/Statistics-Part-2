1. What is hypothesis testing in statistics?
Answer: Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample of data to infer that a certain condition holds true for the entire population.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
Answer: The null hypothesis (H₀) is a statement that there is no effect or no difference, and it serves as the default or starting assumption. The alternative hypothesis (H₁ or Ha) is what you want to prove—it suggests there is an effect or a difference.



3. What is the significance level in hypothesis testing, and why is it important?
Answer: The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common levels are 0.05, 0.01, and 0.10. It determines the threshold for statistical significance.

4. What does a P-value represent in hypothesis testing?
Answer: The P-value is the probability of obtaining test results at least as extreme as the observed results, under the assumption that the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
Answer: A small P-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis, so you reject the null. A large P-value suggests weak evidence against the null, so you fail to reject it.

6. What are Type 1 and Type 2 errors in hypothesis testing?
Answer: A Type 1 error occurs when the null hypothesis is wrongly rejected (false positive). A Type 2 error occurs when the null hypothesis is wrongly not rejected (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
Answer: A one-tailed test checks for the possibility of an effect in one direction, while a two-tailed test checks for the possibility of an effect in both directions.

8. What is the Z-test, and when is it used in hypothesis testing?
Answer: A Z-test is used when the population variance is known and the sample size is large (n ≥ 30). It determines if there is a significant difference between sample and population means.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
Answer: The T-distribution is used when the sample size is small (n < 30) and/or the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails.

11. What is the difference between a Z-test and a T-test?
Answer: A Z-test is used when population variance is known and sample size is large; a T-test is used when population variance is unknown and/or sample size is small.

12. What is the T-test, and how is it used in hypothesis testing?
Answer: A T-test is used to determine if there is a significant difference between the means of two groups, especially when sample sizes are small and variances are unknown.

13. What is the relationship between Z-test and T-test in hypothesis testing?
Answer: Both tests compare means, but a Z-test is used for large samples or known population variance, while a T-test is used for smaller samples or unknown variance.

14. What is a confidence interval, and how is it used to interpret statistical results?
Answer: A confidence interval gives a range of values within which the true population parameter is likely to lie, with a certain level of confidence (e.g., 95%).

15. What is the margin of error, and how does it affect the confidence interval?
Answer: The margin of error reflects the range of uncertainty around the sample estimate. A larger margin of error means a wider confidence interval, indicating less precision.

16. How is Bayes’ Theorem used in statistics, and what is its significance?
Answer: Bayes’ Theorem calculates the probability of a hypothesis given observed data. It's important in updating probabilities based on new evidence.

17. What is the Chi-square distribution, and when is it used?
Answer: The Chi-square distribution is used in tests of independence and goodness of fit. It's applied to categorical data to see if distributions differ from expected patterns.

18. What is the Chi-square goodness of fit test, and how is it applied?
Answer: This test compares observed categorical data to a theoretical distribution. It evaluates whether a sample matches a population with a known distribution.

19. What is the F-distribution, and when is it used in hypothesis testing?
Answer: The F-distribution is used primarily in ANOVA and tests that compare two variances. It's skewed and only takes positive values.

20. What is an ANOVA test, and what are its assumptions?
Answer: ANOVA (Analysis of Variance) tests whether there are significant differences among three or more group means. Assumptions include normality, homogeneity of variance, and independence.

21. What are the different types of ANOVA tests?
Answer: Common types include one-way ANOVA (one independent variable), two-way ANOVA (two independent variables), and repeated measures ANOVA.


22. What is the F-test, and how does it relate to hypothesis testing?
Answer: The F-test compares variances to determine if group means differ significantly. It’s used in ANOVA and regression analysis.



```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

# Known population parameters
population_mean = 100
population_std = 15

# Simulated sample data
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=30)

# Calculate sample mean and size
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Perform Z-test
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z_score, 4))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

import numpy as np
from scipy import stats

# Simulate random sample data
np.random.seed(123)
sample = np.random.normal(loc=50, scale=10, size=40)  # mean=50, std=10, n=40

# Population mean (assumed)
population_mean = 52

# Calculate sample mean and standard deviation
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
sample_size = len(sample)

# Perform one-sample T-test
t_statistic, p_value = stats.ttest_1samp(sample, population_mean)

# Output results
print("Sample Mean:", round(sample_mean, 2))
print("T-statistic:", round(t_statistic, 4))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parameters
population_mean = 100
population_std = 15
alpha = 0.05

# Sample
np.random.seed(101)
sample = np.random.normal(loc=105, scale=15, size=36)
sample_mean = np.mean(sample)
sample_size = len(sample)

# Z-score and p-value
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Plot
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x < -norm.ppf(1 - alpha/2)), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, y, where=(x > norm.ppf(1 - alpha/2)), color='red', alpha=0.3)
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z = {z_score:.2f}')
plt.title('Two-tailed Z-Test Decision Regions')
plt.legend()
plt.show()

# Output
print("Z-score:", round(z_score, 4))
print("P-value:", round(p_value, 4))
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=0, mu1=1, sigma=1, alpha=0.05):
    x = np.linspace(-4, 6, 1000)
    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)
    critical_value = norm.ppf(1 - alpha)

    plt.plot(x, y0, label='Null Hypothesis (H0)', color='blue')
    plt.plot(x, y1, label='Alternative Hypothesis (H1)', color='green')
    plt.fill_between(x, y0, where=(x > critical_value), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=(x <= critical_value), color='orange', alpha=0.3, label='Type II Error')
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')
    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.show()

visualize_errors()
```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

# Simulated data for two independent groups
np.random.seed(123)
group1 = np.random.normal(70, 10, 30)
group2 = np.random.normal(75, 12, 30)

# Perform independent T-test
t_stat, p_value = ttest_ind(group1, group2)

# Output
print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two groups.")
```

```python
# 7. Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

# Simulate before and after data (paired)
np.random.seed(42)
before = np.random.normal(65, 8, 25)
after = before + np.random.normal(2, 4, 25)

# Perform paired T-test
t_stat, p_value = ttest_rel(before, after)

# Visualization
plt.plot(before, label='Before', marker='o')
plt.plot(after, label='After', marker='o')
plt.title('Before vs After (Paired Data)')
plt.legend()
plt.show()

# Output
print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference before and after.")
else:
    print("Fail to reject the null hypothesis: No significant difference before and after.")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    margin = t.ppf((1 + confidence) / 2., n-1) * std_err
    lower = mean - margin
    upper = mean + margin
    return (round(lower, 2), round(upper, 2))

# Simulate sample
np.random.seed(77)
sample = np.random.normal(60, 8, 35)

# Calculate confidence interval
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")

# Explanation
print("This means we are 95% confident that the true population mean lies within this interval.")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import norm

# Simulated sample
np.random.seed(5)
sample = np.random.normal(60, 10, 50)

# Sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# Confidence level
confidence = 0.95
z_score = norm.ppf((1 + confidence) / 2)
margin_of_error = z_score * (sample_std / np.sqrt(n))

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Margin of Error (95% confidence): ±{margin_of_error:.2f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Example: Disease testing
# P(Disease) = 0.01, P(Positive|Disease) = 0.99, P(Positive|No Disease) = 0.05

p_disease = 0.01
p_positive_given_disease = 0.99
p_positive_given_no_disease = 0.05
p_no_disease = 1 - p_disease

# Bayes' Theorem
p_positive = (p_positive_given_disease * p_disease) + (p_positive_given_no_disease * p_no_disease)
p_disease_given_positive = (p_positive_given_disease * p_disease) / p_positive

print(f"Probability of disease given a positive test: {p_disease_given_positive:.4f}")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
data = pd.DataFrame([[20, 15], [30, 35]], columns=['Group A', 'Group B'], index=['Success', 'Failure'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square Statistic:", round(chi2, 4))
print("P-value:", round(p, 4))
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

# Observed data (2x2 table)
observed = np.array([[25, 20], [15, 40]])

# Compute row and column sums
row_sums = observed.sum(axis=1).reshape(-1, 1)
col_sums = observed.sum(axis=0).reshape(1, -1)
total = observed.sum()

# Calculate expected frequencies
expected = row_sums @ col_sums / total

print("Observed Frequencies:\n", observed)
print("Expected Frequencies:\n", expected)
```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

from scipy.stats import chisquare

# Observed and expected frequencies
observed = [16, 18, 16, 14, 12, 24]
total = sum(observed)
expected = [total / 6] * 6  # Equal expected frequencies that sum to 100

# Chi-square test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output
print("Chi-square Statistic:", round(chi2_stat, 4))
print("P-value:", round(p_value, 4))
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples.

from scipy.stats import f

# Simulated data
np.random.seed(101)
group1 = np.random.normal(10, 2, 30)
group2 = np.random.normal(10, 4, 30)

# Calculate F-statistic
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)
f_stat = var1 / var2 if var1 > var2 else var2 / var1

df1 = len(group1) - 1
df2 = len(group2) - 1

p_value = 1 - f.cdf(f_stat, df1, df2)

print("F-statistic:", round(f_stat, 4))
print("P-value:", round(p_value, 4))
```

```python
# 17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

import numpy as np
from scipy.stats import f_oneway

# Simulate data for 3 groups
np.random.seed(202)
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)
group3 = np.random.normal(48, 5, 30)

# One-way ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", round(f_stat, 4))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference among group means.")
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    print("Normality Test (Shapiro-Wilk):")
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        print(f"Group {i+1}: p-value = {p:.4f}")
    stat, p = levene(*groups)
    print("\nEqual Variance Test (Levene's): p-value =", round(p, 4))

check_anova_assumptions(group1, group2, group3)
```

```python
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Simulate dataset
np.random.seed(303)
df = pd.DataFrame({
    'Score': np.random.normal(50, 10, 60),
    'FactorA': np.repeat(['Low', 'High'], 30),
    'FactorB': np.tile(['A', 'B', 'C'], 20)
})

# Two-way ANOVA
model = ols('Score ~ C(FactorA) + C(FactorB) + C(FactorA):C(FactorB)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

from scipy.stats import chi2

# Simulate data
sample = np.random.normal(50, 5, 30)
sample_var = np.var(sample, ddof=1)
hypothesized_var = 25
df = len(sample) - 1

chi2_stat = (df * sample_var) / hypothesized_var
p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

print("Chi-square statistic:", round(chi2_stat, 4))
print("P-value:", round(p_value, 4))
```
